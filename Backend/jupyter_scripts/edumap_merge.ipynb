# Transform AFD_MUNICIPIOS_2017.xlsx

```python
docent_group = pd.read_excel('AFD_MUNICIPIOS_2017.xlsx', skiprows=6)
docent_group
```

```python
pattern1 = 'Total'
id1 = [i for i, e in enumerate(docent_group['LocalizaÃ§Ã£o']) if e == pattern1]

pattern2 = 'PÃºblica'
id2 = [i for i, e in enumerate(docent_group['DependÃªncia Administrativa'].iloc[id1]) if e == pattern2]

docent_group.iloc[id1].iloc[id2]
```

```python
import numpy as np 

col = ['CÃ³digo do MunicÃ­pio', 'Unnamed: 27', 'Unnamed: 28', 'Unnamed: 29', 'Unnamed: 30', 'Unnamed: 31']
docent_group_clean = docent_group[col].iloc[id1].iloc[id2]

docent_group_clean.columns = ['CO_MUNICIPIO', 'Grupo 1', 'Grupo 2', 'Grupo 3', 'Grupo 4', 'Grupo 5']
docent_group_clean['CO_MUNICIPIO'] = [np.int64(e) for e in docent_group_clean['CO_MUNICIPIO']]
docent_group_clean
```
